```python
import torch
import cv2
import numpy as np
from ultralytics import YOLO
model = YOLO('yolov8m.pt')
```

```python
# Train the model
model.train(
    data="/home/hiren/Desktop/Survelliance/weapon-detection/custom-data.yaml",  # Path to dataset YAML
    epochs=20,         # Number of training epochs
    batch=4,          # Batch size
    imgsz=416,         # Image size
    device="cpu"      # Use "cpu" if no GPU is available
)
```

```python
import cv2
from ultralytics import YOLO

# Load the trained YOLO model
model = YOLO("/home/hiren/Desktop/Weapon-Detection/best (1).pt")

# Open webcam (0 = default webcam)
cap = cv2.VideoCapture(0)

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break  # Exit loop if webcam is unavailable

    # Perform object detection
    results = model.predict(frame, conf=0.5)

    # Annotate frame with detections
    annotated_frame = results[0].plot()

    # Display the frame
    cv2.imshow("Live Weapon Detection", annotated_frame)

    # Press 'q' to exit the webcam stream
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release resources
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
from ultralytics import YOLO

# Load the trained YOLO model
model = YOLO("/home/hiren/Desktop/Weapon-Detection/best (1).pt")

# Open the video file
video_path = "/home/hiren/Desktop/Weapon-Detection/videotest.mp4"
cap = cv2.VideoCapture(video_path)

# Get video properties
fps = int(cap.get(cv2.CAP_PROP_FPS))
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

# Define video writer to save output
output_path = "//home/hiren/Desktop/Weapon-Detection/knife_detection_output.mp4"
fourcc = cv2.VideoWriter_fourcc(*"mp4v")  # Codec for saving
out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break  # Exit loop if video ends

    # Perform object detection on the frame
    results = model.predict(frame, conf=0.5)  # Confidence threshold

    # Draw bounding boxes on the frame
    annotated_frame = results[0].plot()

    # Show the frame (optional)
    cv2.imshow("Weapon Detection", annotated_frame)
    
    # Write the frame to output video
    out.write(annotated_frame)

    # Press 'q' to stop early
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release everything
cap.release()
out.release()
cv2.destroyAllWindows()
```
